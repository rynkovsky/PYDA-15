# Метрики precision и recall

Данные содержат признаки пользователей и целевую переменную affair - была ли связь на стороне

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings

warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('E:\\Python\\AML\\HW 3\\affair_data.csv')
data.head()
```

```python
data.info()
```

```python
from sklearn.linear_model import LogisticRegression

# импортируем метод для автоматической разбивки на обучающую и тестовую выборки
# раньше использовали from sklearn.cross_validation import train_test_split  
from sklearn.model_selection import train_test_split
```

Формируем набор признаков и целевую переменную

```python
X = data[data.columns[:-1]]
X.head(2)
```

```python
Y = data['affair']
```

```python
model = LogisticRegression()
```

Разбиваем данные на обучающую и тестовую выборки в соотношении 70 / 30

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, 
                                                    test_size=0.3, 
                                                    random_state=0)
```

Обучаем модель

```python
model.fit(X_train, 
          Y_train)
```

Получаем прогнозные значения модели (переменная predictions)

```python
predictions = model.predict_proba(X_test)
predictions
```

```python
model.score(X_test, Y_test)
```

Построим набор моделей для значений random_state от 0 до 9. Получите model.score для каждого варианта.

```python
for i in range(10):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, 
                                                        test_size=0.3, 
                                                        random_state=i)
    model = LogisticRegression()
    model.fit(X_train, Y_train)
    print("{}: {}".format(i, model.score(X_test, Y_test)))
```

### Оценка качества модели

Считаем accuracy последней модели

```python
tp = 0 # True positive
fp = 0 # False positive
fn = 0 # False negative
tn = 0 # True negative

predictions = model.predict_proba( X_test )
for predicted_prob, actual in zip( predictions[:, 1], Y_test ):    
    if predicted_prob >= 0.5:
        predicted = 1
    else:
        predicted = 0
    
    if predicted == 1:
        if actual == 1:
            tp += 1
        else:
            fp += 1
    
    else:
        if actual == 1:
            fn += 1
        else:
            tn += 1
```

```python
# какая точность получилась?
(tp + tn) / (tp + fp + fn + tn)
```

```python
precision = tp / (tp + fp)
precision
```

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score
```

```python
preds = model.predict(X_test)
accuracy_score(Y_test, preds)
```

```python
precision_score(Y_test, preds), recall_score(Y_test, preds)
```

### Используем готовые библиотеки

```python
# метод для построения графика precision-recall
from sklearn.metrics import precision_recall_curve
```

```python
pr, re, thres = precision_recall_curve( Y_test, predictions[:, 1] )
```

```python
# посмотрим что получилось
for line in zip( pr, re, thres ):
    print( line )
```

```python
from sklearn.metrics import average_precision_score
```

```python
# площадь под кривой
average_precision_score( Y_test, predictions[:, 1] )
```

### ROC

```python
# посчитаем ROC-кривую (Receiver Operating Characteristic)
from sklearn.metrics import roc_curve
```

```python
# получаем значения false и true positive rate для различных значений порога
fpr, tpr, thres = roc_curve( Y_test, predictions[:, 1] )
```

```python
from sklearn.metrics import auc
```

```python
import numpy as np
plt.figure(figsize=(12,6))
plt.title("ROC Curve", fontsize=18)
plt.grid()
plt.plot(fpr, tpr, label='classifier predictions')
plt.plot(np.linspace(0,1,100), 
         np.linspace(0,1,100), 
         linestyle='--', 
         label='random predictions')
plt.legend(fontsize=14)
plt.show()
print("AUC is %.2f" %auc(fpr, tpr))
```

Посчитать tpr и fpr в ноутбуке к лекции. Убедиться, что график ROC AUC получается таким же, как и штатными средствами.

Построить график Precision - Recall на этих же данных.

```python
import numpy as np

tp = 0 # True positive
fp = 0 # False positive
fn = 0 # False negative
tn = 0 # True negative

fpr3 = []
tpr3 = []
precision3 = []

predictions = model.predict_proba( X_test )

for threshold in np.linspace(0, 1, 10): 
           
    for predicted_prob, actual in zip( predictions[:, 1], Y_test ):  
        
        if predicted_prob >= threshold:
            predicted = 1
        else:
            predicted = 0

        if predicted == 1:
            if actual == 1:
                tp += 1
            else:
                fp += 1

        else:
            if actual == 1:
                fn += 1
            else:
                tn += 1

    fpr3.append(fp / (fp + tn))
    tpr3.append(tp / (tp + fn))
    precision3.append(tp / (tp + fp))
    
fpr4 = np.asarray(fpr3)
tpr4 = np.asarray(tpr3)
precision4 = np.asarray(precision3)
recall4 = tpr4      
```

```python
import numpy as np
plt.figure(figsize=(12,6))
plt.title("ROC Curve", fontsize=18)
plt.grid()
plt.plot(fpr4, tpr4, label='classifier predictions')
plt.plot(np.linspace(0,1,100), 
         np.linspace(0,1,100), 
         linestyle='--', 
         label='random predictions')
plt.xlabel('False Positive Rate', fontsize=14)
plt.ylabel('True Positive Rate', fontsize=14)
plt.legend(fontsize=14)
plt.show()
print("AUC is %.2f" %auc(fpr4, tpr4))
```

```python

plt.figure(figsize=(12,6))
plt.title("Precision-Recall Curve", fontsize=18)
plt.grid()
plt.plot(precision4, recall4, label='classifier predictions')
plt.plot(np.linspace(0,1,100), 
         np.linspace(0,1,100), 
         linestyle='--', 
         label='random predictions')
plt.xlabel('Recall', fontsize=14)
plt.ylabel('Precision', fontsize=14)
plt.legend(fontsize=14)
plt.show()
print("AUC is %.2f" %auc(precision4, recall4))
```
